```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def reverse_in_groups(head, k):
    if not head or k <= 1:
        return head

    # Function to reverse a linked list within a given range
    def reverse_range(start, end):
        prev, current = None, start
        while current != end:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        return prev

    # Helper function to find the end of the reversed group
    def find_group_end(node, group_size):
        for _ in range(group_size - 1):
            if not node:
                return None
            node = node.next
        return node

    num_nd = ListNode(0)
    num_nd .next = head
    prev_group_end = num_nd
    current = head

    while current:
        group_end = find_group_end(current, k)
        if not group_end:
            break

        next_group_start = group_end.next
        group_end.next = None

        prev_group_end.next = reverse_range(current, next_group_start)
        current.next = next_group_start

        prev_group_end = current
        current = next_group_start

    return num_nd.next

def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> ")
        current = current.next
    print("None")

def create_linked_list():
    values = list(map(int, input("Enter space-separated values for the linked list: ").split()))
    nodes = [ListNode(value) for value in values]
    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i + 1]
    return nodes[0]

head = create_linked_list()
group_size = int(input("Enter the group size for reversing: "))

# Reverse the linked list in groups
new_head = reverse_in_groups(head, group_size)

# Print the reversed linked list
print("Reversed Linked List:")
print_linked_list(new_head)
```
